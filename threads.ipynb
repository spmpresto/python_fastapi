```python
from threading import Thread
from time import perf_counter, sleep
import requests
```

## Декоратор для подсчета времени

```python
def timeit(func):
    def wrapper(*args, **kwargs):
        start = perf_counter()
        result = func(*args, **kwargs)
        execution_time = perf_counter() - start
        print(execution_time)
        return result
    return wrapper
```

```python
@timeit
def power(num1, times):
    for _ in range(times):
        num1 * num1
    return "Готово"
```

# CPU bound

```python
big_number = 1_000_000
power(big_number, big_number)
```

```python
threads = []
for i in range(100):
    thread = Thread(target=power, args=(big_number, big_number))
    threads.append(thread)

for thread in threads:
    thread.start()

for thread in threads:
    thread.join()
```

# IO bound

```python
@timeit
def make_api_call(id):
    sleep(1)
    return requests.get(f"https://jsonplaceholder.typicode.com/todos/{id}").json()
```

```python
make_api_call(1)
```

```python
threads = []
for id in range(10):
    thread = Thread(target=make_api_call, args=(id,))
    threads.append(thread)

for thread in threads:
    thread.start()

for thread in threads:
    thread.join()
```
