```python
import asyncio
from time import perf_counter
import aiohttp
```

```python
def timeit(func):
    async def wrapper(*args, **kwargs):
        start = perf_counter()
        result = await func(*args, **kwargs)
        execution_time = perf_counter() - start
        print(execution_time)
        return result
    return wrapper
```

```python
@timeit
async def make_api_call(id):
    await asyncio.sleep(1)  
    async with aiohttp.ClientSession() as session:
        async with session.get(f"https://jsonplaceholder.typicode.com/todos/{id}") as resp:
            return await resp.json()
```

```python
await make_api_call(1)
await make_api_call(2)
await make_api_call(3)
```

```python
async def make_miltiple_requests():
    for i in range(1, 6):
        print(f"Отправлен запрос id={i}")
        await make_api_call(i)
```

```python
await make_miltiple_requests()
```

## Одновременное выполнение асинхронных функций

```python
result = await asyncio.gather(
    make_api_call(1),
    make_api_call(2),
    make_api_call(3),
    make_api_call(4),
    make_api_call(5),
)
result
```
